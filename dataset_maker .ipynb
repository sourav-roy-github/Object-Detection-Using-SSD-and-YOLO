```python
import shutil
import random

from os import listdir
from os.path import isfile, join

random.seed(1)
```

```python
dataset_dir = "sdcar-dataset/export"
```

```python
images_dir = dataset_dir + "/images"
labels_dir = dataset_dir + "/labels"
```

```python
image_files = [f for f in listdir(images_dir) if isfile(join(images_dir, f))]
```

```python
image_files
```

```python
image_part_list = []
unique_images_list = []

for image_name in image_files:
    image_parts = image_name.split(".")
    if image_parts[0] not in image_part_list:
        unique_images_list.append(image_name)
        image_part_list.append(image_parts[0])
        
```

```python
len(unique_images_list)
```

```python
# create test and train directories"
def create_directory(path):
    dir_exists = os.path.exists(path)
    if not dir_exists:
        os.mkdir(path)
```

```python
# create directories

yolo_dataset_dir = "sd_car_dataset_for_yolo"
train_dataset_images_dir = yolo_dataset_dir + "/train"
train_dataset_images_dir = yolo_dataset_dir + "/train"
test_dataset_dir = yolo_dataset_dir + "/train"

dataset_directories_list = [yolo_dataset_dir, 
                            yolo_dataset_dir + "/images",
                            yolo_dataset_dir + "/labels",
                            yolo_dataset_dir + "/images/train", 
                            yolo_dataset_dir + "/images/val",
                            yolo_dataset_dir + "/labels/train",
                            yolo_dataset_dir + "/labels/val"]

for dir_ele in dataset_directories_list:
    create_directory(dir_ele)
```

```python
def test_train_split(images_list, ratio):
    # shuffle list
    random.shuffle(images_list)
    elements = len(images_list)
    middle = int(elements * ratio)
    return [images_list[middle:], images_list[:middle]]
```

```python
train_images, test_images = test_train_split(unique_images_list, 0.2)
```

```python
len(train_images)
```

```python
len(test_images)
```

```python
# copy train images and labels
for train_img in train_images:
    train_label = train_img.strip(".jpg") + ".txt"
    image_path = images_dir + "/" + train_img
    label_path = labels_dir + "/" + train_label
    if os.path.isfile(label_path):
        shutil.copy2(image_path, dataset_directories_list[3])
        shutil.copy2(label_path, dataset_directories_list[-2])
    else:
        print("label missing")
```

```python
# copy test images and labels
for test_img in test_images:
    test_label = test_img.strip(".jpg") + ".txt"
    image_path = images_dir + "/" + test_img
    label_path = labels_dir + "/" + test_label
    if os.path.isfile(label_path):
        shutil.copy2(images_dir + "/" + test_img, dataset_directories_list[-3])
        shutil.copy2(labels_dir + "/" + test_label, dataset_directories_list[-1])
    else:
        print("Label missing")
```
